### Approaches to handle nulls
1) We can remove the rows which are having null values but removing anything from data can affect our insights so may not be the good approach
2) We can impute the values with the mean and median. Going ahead with mean is not a preferable choice as mean is affected by outliers so can introduce a bias in our dataset. Median is more robust as it is the count of freequency of an item existing multiple times and is immune to outliers.

But here we have different categories of data so directly taking median of all of them together will also bring some bias and reviews rating is very important insights. So we will find the median for each category and then impute it with the missing values of each category

```python
df['Review Rating']=df.groupby('Category')["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
#create a column age group
labels=['Young_Adult','Adult','Middle-aged','Senior'] #this is a list of all the different categories
df['age_group']=pd.qcut(df['age'],q=4,labels=labels) #this is a way we are introducing new column
```

### Qcut and its explanantion

Sort all values in df['age']

Split them into 4 equal-sized groups

Each group has ~25% of the rows

Assign labels in order:

Lowest 25% values ‚Üí Young_Adult

Next 25% ‚Üí Adult

Next 25% ‚Üí Middle-aged

Highest 25% ‚Üí Senior

While we can go ahead with this but this might fail if our min and max range in the age is between 18-30 so 30 can't be considered senior so for that we should take pd.cut
